## Purchasing Analysis (Total)

```python
#calculations for unique items, average price,number of purchases, total revenue
purchase_file_df.head()
number_of_unique_items=len(purchase_file_df["Item ID"].unique())
average_price=purchase_file_df["Price"].mean()
number_of_purchases=purchase_file_df["Purchase ID"].count()
total_revenue=purchase_file_df["Price"].sum()

#create summary data frame
summary_data_items= {"Number of Unique Items":[number_of_unique_items],
                "Average Price":[average_price],
                "Number of Purchases":[number_of_purchases],
                "Total Revenue":[total_revenue]}
summary_df=pd.DataFrame(summary_data_items,columns=["Number of Unique Items","Average Price", 
                                                    "Number of Purchases","Total Revenue"])
                            
# use map to format all columns
summary_df['Average Price'] = summary_df['Average Price'].map("${:,.2f}".format)
summary_df['Total Revenue'] = summary_df['Total Revenue'].map("${:,.2f}".format)
summary_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#create a new dataframe grouped by gender
grouped_purchase_file_df = purchase_file_df.groupby(["Gender"])
#print(grouped_purchase_file_df)

#caulculate purchase count 
gender_purchase_count=grouped_purchase_file_df["Purchase ID"].count()

#average price
gender_avg_price=grouped_purchase_file_df["Price"].mean()

#total
gender_total=grouped_purchase_file_df["Price"].sum()

#calculate average total per person
gender_unique_player=grouped_purchase_file_df["SN"].nunique()

#create dataframe
gender_analysis=pd.DataFrame({"Purchase Count":gender_purchase_count,
              "Average Purchase Price":gender_avg_price,
              "Total Purchase Value":gender_total,
              "Average Purchase per Person":gender_total/gender_unique_player})
gender_analysis

#formatting 
gender_analysis_formated=gender_analysis

gender_analysis_formated['Average Purchase Price']=gender_analysis_formated['Average Purchase Price'].map("${:,.2f}".format)
gender_analysis_formated['Total Purchase Value'] =gender_analysis_formated['Total Purchase Value'].map("${:,.2f}".format)
gender_analysis_formated['Average Purchase per Person'] =gender_analysis_formated['Average Purchase per Person'].map("${:,.2f}".format)
gender_analysis_formated
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#create age bins
age_bins=[0,9,14,19,24,29,34,39,150]

#create age labels
age_labels=["<10","10-14","15-19","20-24","25-29","30-34","35-39",">40"]

purchase_file_df["Age Range"]=pd.cut(x=purchase_file_df['Age'], bins=age_bins, labels=age_labels)
purchase_file_df.head(15)

#group data by age
groupedage_purchase_file = purchase_file_df.groupby(["Age Range"])

#calculate player count
player_count=groupedage_purchase_file["SN"].nunique()
player_count

#calculate player percentage
player_percentage=100*player_count/player_count.sum()
player_percentage

#create data frame
age_analysis=pd.DataFrame({"Player Count":player_count,
              "Player Percentage":player_percentage})
#age_analysis

#Format dataframe
age_analysis_updated = age_analysis

age_analysis_updated['Player Percentage'] = age_analysis_updated['Player Percentage'].map("{:.2f}%".format)
age_analysis_updated
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#group data by Screen Name

spender_data=purchase_file_df.groupby(["SN"])
#spender_data.head()

#Calculate purchase count
spender_purchase_count=spender_data["Purchase ID"].count()

#Calculate Avg.purchase price
spender_avg_price=spender_data["Price"].mean()

#calculate total purchase value
spender_total_purchase=spender_data["Price"].sum()

#calculate avg total per person
#unique_spender=spender_data["SN"].nunique()

#create data frame
leader_spenders=pd.DataFrame({"Purchase Count":spender_purchase_count,
                              "Average Purchase Price":spender_avg_price,
                              "Total Purchase Value":spender_total_purchase,})

#sort top spenders
leader_spenders_reviewed=leader_spenders.sort_values(["Total Purchase Value"],ascending=False)
leader_spenders_reviewed

#formatting

leader_spenders_reviewed['Average Purchase Price']=leader_spenders_reviewed['Average Purchase Price'].map("${:,.2f}".format)
leader_spenders_reviewed['Total Purchase Value'] = leader_spenders_reviewed['Total Purchase Value'].map("${:,.2f}".format)
leader_spenders_reviewed


#leader_spender_reviewed['Average Purchase Price'] = leader_spender_reviewed['Average Purchase Price'].map("${:,.2f}"
leader_spenders_reviewed.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Create a new dataframe with required data
#demographics=purchase_file_df[["SN","Gender","Age"]]
#popular_data=purchase_file_df[["Item ID","Item Name","Price"]]

#Group items data by Item ID and Item Name
#spender_data = purchase_file_df.groupby(["SN"])
popular_data=purchase_file_df.groupby(["Item ID","Item Name"])


#Calculate popular items purchase count
popular_purchase_count=popular_data["Purchase ID"].count()
popular_purchase_count

#calculate total purchase value
popular_purchase_value=popular_data["Price"].sum()
popular_purchase_value

#Calculate Item Price
popular_item_price=popular_purchase_value/popular_purchase_count
popular_item_price

#Create dataframe to display the calculations
popular_items=pd.DataFrame({"Purchase Count":popular_purchase_count,
              "Item Price":popular_item_price,
              "Total Purchase Value":popular_purchase_value,})
popular_items
#sort purchase count column in descending order
popular_items_updated=popular_items.sort_values(["Purchase Count"],ascending=False)
popular_items_updated
#formatting
format_popular_items_updated=popular_items_updated
format_popular_items_updated['Item Price']=format_popular_items_updated['Item Price'].map("${:,.2f}".format)
format_popular_items_updated['Total Purchase Value'] = format_popular_items_updated['Total Purchase Value'].map("${:,.2f}".format)
format_popular_items_updated.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#Sort the above table by total purchase value in descending order

most_profitable_items=popular_items.sort_values(["Total Purchase Value"],ascending=False)
most_profitable_items

#formatting
formatted_most_profitable_items=most_profitable_items
formatted_most_profitable_items['Item Price'] = formatted_most_profitable_items['Item Price'].map("${:,.2f}".format)
formatted_most_profitable_items['Total Purchase Value'] = formatted_most_profitable_items['Total Purchase Value'].map("${:,.2f}".format)
formatted_most_profitable_items.head()
```
